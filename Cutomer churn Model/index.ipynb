```python
x = df.drop(columns=["Exited"])
y = df["Exited"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.20, random_state=42, stratify=y
)
```

#### Standard Scalar or One hot encoding

```python
numeric = df.select_dtypes(exclude=["object"]).columns
catgorical = df.select_dtypes(include=["object"]).columns
```

```python
numeric = [
    "CreditScore",
    "Age",
    "Tenure",
    "Balance",
    "NumOfProducts",
    "HasCrCard",
    "IsActiveMember",
    "EstimatedSalary",
]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
x_train.columns
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
pipeline = Pipeline([("column_transform", column_transform), ("log", lr)])
```

```python
pipeline.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
y_pred = pipeline.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
```

```python
lor = LogisticRegression()
svc = SVC(kernel="sigmoid", gamma=0.1)
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
bagg = BaggingClassifier()
ada = AdaBoostClassifier()
etc = ExtraTreesClassifier()
gra = GradientBoostingClassifier()
xgb = XGBClassifier()
```

```python
clfs = {
    "lor": lor,
    "SVC": svc,
    "knn": knn,
    "dt": dt,
    "rf": rf,
    "bagg": bagg,
    "ada": ada,
    "etc": etc,
    "gra": gra,
    "xgb": xgb,
}
```

```python
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score
from sklearn.model_selection import GridSearchCV
```

```python
param_grids = {
    "lor": { 
        "model__C": [0.1, 1, 10], 
        "model__solver": ["liblinear", "lbfgs"]
    },
    "SVC": {
         "model__C": [0.1, 1, 10], 
         "model__kernel": ["linear", "rbf"]
    },
    "knn": {
         "model__n_neighbors": [3, 5, 7], 
         "model__weights": ["uniform", "distance"]
    },
    "dt": {
        "model__max_depth": [None, 10, 20, 30],
        "model__min_samples_split": [2, 5, 10],
    },
    "rf": {
        "model__n_estimators": [50, 100, 200],
        "model__max_depth": [None, 10, 20],
        "model__min_samples_split": [2, 5, 10],
    },
    "bagg": {"model__n_estimators": [50, 100, 200]},
    "ada": {
        "model__n_estimators": [50, 100, 200],
        "model__learning_rate": [0.01, 0.1, 1],
    },
    "etc": {
         "model__n_estimators": [50, 100, 200], 
         "model__max_depth": [None, 10, 20]
    },
    "gra": {
        "model__n_estimators": [50, 100, 200],
        "model__learning_rate": [0.01, 0.1, 1],
    },
    "xgb": {
        "model__n_estimators": [50, 100, 200],
        "model__learning_rate": [0.01, 0.1, 1],
        "model__max_depth": [3, 6, 9],
    },
}
```

```python
def Train_Model(clf, param_grid, x_train, y_train, x_test, y_test):

    pipeline = Pipeline(steps=[("column_tran", column_transform), ("model", clf)])

    grid_search = GridSearchCV(
        estimator=pipeline, param_grid=param_grid, cv=10, scoring="accuracy"
    )
    grid_search.fit(x_train, y_train)

    best_pipeline = grid_search.best_estimator_
    y_pred = best_pipeline.predict(x_test)

    acc = accuracy_score(y_test, y_pred)
    ps = precision_score(y_test, y_pred, average="weighted")
    rec = recall_score(y_test, y_pred, average="weighted")
    f1 = f1_score(y_test, y_pred, average="weighted")
    return acc, ps, rec, f1, grid_search.best_params_
```

```python
acc_score = []
pre_score = []
recall = []
f1s = []

for key, value in clfs.items():
    param_grid = param_grids.get(key)
    acc, ps, rs, f1 , best_params= Train_Model(value, param_grid, x_train, y_train, x_test, y_test)

    print("name:", key)
    print("ac:", acc)
    print("ps :", ps)
    print("rs:", rs)
    print("f1:", f1)
    print("Best params : ",best_params)

    acc_score.append(acc)
    pre_score.append(ps)
    recall.append(rs)
    f1s.append(f1)
```

```python
model_df = pd.DataFrame(
    {
        "Algoritham": clfs.keys(),
        "Accuracy Score": acc_score,
        "Precision Score": pre_score,
        "Recall Score": recall,
        "F1 Score": f1s,
    }
).sort_values("Accuracy Score", ascending=False)
model_df
```

### Train the XGB model

```python
def xgb_model(clf,param_grid,x_train,y_train):
     xgb_pipeline = Pipeline(steps=[
          ('column_tra',column_transform),
          ('model',clf)
     ])

     grid_search = GridSearchCV(estimator=xgb_pipeline,param_grid=param_grid,cv=5,scoring='accuracy')
     grid_search.fit(x_train,y_train)

     best_pipeline = grid_search.best_estimator_

     return best_pipeline,grid_search.best_params_
```

```python
xgb_parameter_grid = param_grids['xgb']
```

```python
best_model , best_params = xgb_model(clfs['xgb'],xgb_parameter_grid,x_train,y_train)
```
